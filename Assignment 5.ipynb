```python
n, m = 30, 10000
x = stats.norm(loc=0, scale=1).rvs(size=n)

with pm.Model() as normal_gamma_prior:
    theta0 = 1 
    tau0 = 1
    theta = pm.Normal("theta", mu=theta0, sigma=tau0**(-0.5))

    tau_SS_prior = 1  
    tau_prior_n = 1 
    tau = pm.Gamma("tau", alpha=tau_prior_n/2, 
                            beta=tau_SS_prior/2)

    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=x)
    idata = pm.sample(draws=m, chains=2)
```

$\textbf{Assessments:}$ The Bayesian inference posterior produced by normal-gamma prior has similar performance to the inference from Gibbs Sampling, having the similar range of the intervals, and slightly better metropolis within Gibbs depending the values of hyperparameters.

#### Second Model

```python
with pm.Model() as non_normal_gamma_prior1:
    theta = pm.Cauchy("theta", alpha=1, beta=1)

    tau = pm.Exponential("tau", lam=2)

    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=x)
    idata = pm.sample(draws=m, chains=2)
```

Third Model:

The priors:

$p(\theta) \sim  \text{Laplace}(\mu = 1, b = 2)$, \

### Q3

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):
    
    # find interval of grid points where f(x_grid) > y 
    # then extend the enterval so endpoints f(a)<y and f(b)<y 
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid
    
    x_ = a + stats.uniform().rvs()*(b-a)    
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f" 
    elif x_ < x:  # or if "x_ was above f on the left side of the interval" 
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):
    
    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)
    
    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

```python
x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))
    
m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25
plot_trace = 10
for t in range(1,m):
    
    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')
    
    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')

plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```

In this code, the sample from $\text{Beta}(\alpha=1.5, \beta=3)$ is drawn. The blue curve is the true underlying pdf, and the orange histogram is the sample through slice sampling. 

The algorithm proceed as the following: Suppose we want to sample from the target distribution with a known probability density function $f(x)$.
1. Start with the initial value $x^{(0)}$ such that $f(x^{(0)}) > 0$.
2. For t from 1 to m: \
&emsp; - Sample $y^{(t)}$ from $\text{Uniform}(0, x^{(t-1)})$. \
&emsp; - Use $y^{(t)}$ to "slice" the true p.d.f. $f(x)$ by only sample $x^{(t)}$ such that $f(x^{(t)}) > y^{(t)}$. \
&emsp; - (For precision purposes, which is included in this code snippet) if the sampled $x^{(t)}$ does not satisfy $f(x^{(t)}) > y^{(t)}$, then we narrow down the interval, and then sample $x^{(t)}$ again (in `slice_f_at_y_`).
3. The value $x^{(0)}, x^{(1)}, \ldots, x^{(m)}$ is the sample from $f(x)$.

The solid black line in the graph indicates the transition of y from previous iteration to the next iteration $(y^{(t-1)} \rightarrow y^{(t)})$, and the dotted line is the transition of x value $(x^{(t)} \rightarrow x^{(t+1)})$.

We can use slice sampling in the place of Metropolis-Hastings algorithm in Metropolis within Gibbs step, since both of these algorithms produce a stationary distribution. Replacing will get a new algorithm called Slice within Gibbs Sampling, in which slice sampling will be used to sample from the full conditionals. The important thing to note is that the initial value should be the value such that $f(x^{(0)} > 0)$. Also, observe that we do not need to find any proposal distributions in this algorithm, unlinke MH algorithm. The sketch of the algorithm is:
1. Start with the initial value $(\theta_0^{(0)}, \theta_1^{(0)}, \ldots, \theta_n^{(0)})$.
2. For t from 1 to m: \
&emsp; - Sample from the full conditional $\theta_0^{(t)} \sim p(\theta_0|\textbf{x}, \theta_1^{(t-1)}, \ldots, \theta_n^{(t-1)})$. \
&emsp; - Sample from the full conditional $\theta_1^{(t)} \sim p(\theta_0|\textbf{x}, \theta_0^{(t)}, \theta_2^{(t-1)} \ldots, \theta_n^{(t-1)})$. \
&emsp;&emsp;&emsp; $\vdots$ \
&emsp; - Sample from the full conditional $\theta_n^{(t)} \sim p(\theta_0|\textbf{x}, \theta_0^{(t)}, \ldots, \theta_{n-1}^{(t)})$. \
&emsp; Observe that we can use the slice sampling to sample from the full conditional distributions above, give that the p.d.f. o the full conditionals are known.
